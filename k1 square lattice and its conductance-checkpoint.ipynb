```python
lat= kwant.lattice.square(1)
```

```python
t = 1.0
W = 10
L = 30

# Define the scattering region

for i in range(L):
    for j in range(W):
        # On-site Hamiltonian
        syst[lat(i, j)] = 4* t* sigma_0 + e_z * sigma_z

        # Hopping in y-direction
        if j > 0:
            syst[lat(i, j), lat(i, j - 1)] = -t

        # Hopping in x-direction
        if i > 0:
            syst[lat(i, j), lat(i - 1, j)] = -t
```

```python
kwant.plot(syst)
```

```python
sym_left_lead = kwant.TranslationalSymmetry((-1, 0))
left_lead = kwant.Builder(sym_left_lead)
```

```python
for j in range(W):
    left_lead[lat(0, j)] = 4* t* sigma_0 + e_z * sigma_z
    if j > 0:
        left_lead[lat(0, j), lat(0, j - 1)] = -t
    left_lead[lat(1, j), lat(0, j)] = -t
```

```python
syst.attach_lead(left_lead)
```

```python
sym_right_lead = kwant.TranslationalSymmetry((1, 0))
right_lead = kwant.Builder(sym_right_lead)

for j in range(W):
    right_lead[lat(0, j)] = 4 * t
    if j > 0:
        right_lead[lat(0, j), lat(0, j - 1)] = -t
    right_lead[lat(1, j), lat(0, j)] = -t

syst.attach_lead(right_lead)
```

```python
kwant.plot(syst)
```

```python
sys= syst.finalized()
```

```python
local_dos = kwant.solvers.default.ldos(sys, energy=4)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(local_dos)
plt.show()
```

```python
energies = []
data = []
for ie in range(100):
    energy = ie * 0.01

    # compute the scattering matrix at a given energy
    smatrix = kwant.smatrix(sys, energy)

    # compute the transmission probability from lead 0 to
    # lead 1
    energies.append(energy)
    data.append(smatrix.transmission(1, 0))
```

```python
plt.figure()
plt.plot(energies, data)
plt.xlabel("energy [t]")
plt.ylabel("transmission probability")
plt.show()
```
