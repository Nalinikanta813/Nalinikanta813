```python
tweet= re.sub("^a-zA-Z"," ",str(tweet))
```

```python
tweet # >>> got only text data
```

```python
tweet= tweet.lower() # >>>>> lets convert all to lower case
```

```python
tweet
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords= nltk.corpus.stopwords("english")
stop_list= stopwords + ['rt','url']
tweet= " ".join([word for words in tweet.split(" ") if word not in stop_list])
```

```python
lemmatiser1 = nltk.stem.wordnet.lemmatiser1
tweet = " ".join([lemmatiser1.lemmatize(word) for word in tweet.split(" ")])
```
