```python
import kwant
import matplotlib.pyplot as plt
```

```python
syst= kwant.Builder()
lat= kwant.lattice.square(a=1)

L=30
W=5
t=1

syst[(lat(x, y) for x in range(L) for y in range(W))] = 4* t
kwant.plot(syst)
```

```python
syst[lat.neighbors()] = -1
kwant.plot(syst);
```

```python
lead_symm= kwant.TranslationalSymmetry((-1,0))
lead= kwant.Builder(lead_symm)

lead[(lat(0, y) for y in range(W))] = 4* t
lead[lat.neighbors()] = -1

syst.attach_lead(lead)
syst.attach_lead(lead.reversed())
kwant.plot(syst)
```

```python
sys = syst.finalized()
```

```python
kwant.plotter.bands(lead.finalized());
```

```python
local_dos= kwant.ldos(sys,energy=2)
kwant.plotter.map(sys, local_dos, num_lead_cells=10, cmap= 'inferno');
```

```python
ham = sys.hamiltonian_submatrix()
plt.matshow(ham==0, cmap="gray", interpolation=None)
```

```python
import scipy.linalg as la
eval, evec = la.eigh(ham)

kwant.plotter.map(sys, abs(evec[:,0])**2, cmap = 'inferno');
```

```python
energies = []
t_coef = []
for ie in range(1000):
    energy = ie * 0.01
    scatter_matrix = kwant.smatrix(sys, energy)

    
    energies.append(energy)
    t_coef.append(scatter_matrix.transmission(1, 0))
    
plt.figure()
plt.plot(energies,t_coef)
plt.xlabel("energy")
plt.ylabel("transmission_coefficient")
plt.title("conductance of our system")
plt.show()
```
